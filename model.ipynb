```python
import pandas as pd
import numpy as np
```

```python
fake = pd.read_csv("./Fake.csv")
real = pd.read_csv("./True.csv")
```

```python
fake["target"] = 0
fake.head()
```

```python
real["target"] = 1
```

```python
fake.shape,real.shape
```

```python
merged = pd.concat([real,fake],axis=0)
merged.head()
```

```python
merged = merged.drop(["subject","date","title"],axis=1)
```

```python
merged.isnull().sum()
```

```python
#suffleling the data
merged = merged.sample(frac=1)
merged
```

```python
merged.reset_index(inplace = True)#resetting the index as it was suffled before
merged.drop(["index"],axis=1,inplace=True)
```

```python
merged.head()
```

```python
X = merged["text"]
y = merged["target"]
```

```python
#reduction of noise like 
from sklearn.feature_extraction.text import TfidfVectorizer

vector = TfidfVectorizer(
    stop_words="english",
    lowercase=True
)

X_reduce = vector.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_reduce,y,random_state=10,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(X_train,y_train)
```

```python
LR.score(X_train,y_train)
```
